```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Load data from CSV file
data = pd.read_csv('/workspaces/codespaces-jupyter/data/output.csv')

# Assign variables
x = data['Gulnara'].values
y = data['Jaan'].values
z = data['Indrek'].values

# Define the regression equation (change to fit your data)
def f(x, y):
    return x**2 + y**2

# Create a meshgrid of points to plot the surface
x_min, x_max = x.min(), x.max()
y_min, y_max = y.min(), y.max()
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 10), np.linspace(y_min, y_max, 10))
zz = f(xx, yy)

# Fit the data with linear regression (change to fit your data)
A = np.vstack([x, y, np.ones(len(x))]).T
params, _, _, _ = np.linalg.lstsq(A, z, rcond=None)

# Create a plane using the regression parameters
plane = np.array([params[0]*xx + params[1]*yy + params[2]])

# Plot the data points and the plane
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, c='b', marker='o')
ax.plot_surface(xx, yy, zz, alpha=0.5)
ax.plot_surface(xx, yy, plane.reshape(xx.shape), alpha=0.5)
ax.set_xlabel('Gulnara')
ax.set_ylabel('Jaan')
ax.set_zlabel('Indrek')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA

# Load data from CSV file
data = pd.read_csv("/workspaces/codespaces-jupyter/data/output.csv")

# Extract the x, y, and z columns
x = data["Gulnara"].values
y = data["Jaan"].values
z = data["Indrek"].values

# Combine the x, y, and z columns into a 2D array
X = np.column_stack((x, y, z))

# Perform PCA to reduce the dimensionality of the data to 2D
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Plot the embeddings in 3D space
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_pca[:,0], X_pca[:,1], z, alpha=0.5)
ax.set_xlabel('Gulnara')
ax.set_ylabel('Jaan')
ax.set_zlabel('Indrek')
plt.show()
```
